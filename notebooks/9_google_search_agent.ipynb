```python
from dotenv import load_dotenv  
load_dotenv()
```

```python
from langchain_community.utilities import GoogleSerperAPIWrapper
from langchain_groq import ChatGroq
from langchain.agents import create_agent
from langchain_core.tools import tool
```

```python
llm = ChatGroq(model="openai/gpt-oss-120b")
search = GoogleSerperAPIWrapper()
```

```python
@tool
def google_search(question: str):
    """
    you are a tool that searches the web using google search and returns the results.
    """
    return search.run(question)
```

```python
agent = create_agent(
    model=llm,
    tools=[google_search],
    system_prompt="You are a agent who can search the web using google search"
)
```

```python
question = "what is today tech news?"
```

```python
res = agent.invoke({"messages":[{"role":"user","content":question}]})
```

```python
print(res["messages"][-1].content)
```
