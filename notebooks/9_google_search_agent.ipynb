```python
from dotenv import load_dotenv  
load_dotenv()
```

```python
from langchain_community.utilities import GoogleSerperAPIWrapper
from langchain_groq import ChatGroq
from langchain.agents import create_agent
from langchain_core.tools import tool
```

```python
llm = ChatGroq(model="openai/gpt-oss-120b")
search = GoogleSerperAPIWrapper()
```

```python
@tool
def google_search(question: str):
    """
    you are a tool that searches the web using google search and returns the results.
    """
    return search.run(question)
```

```python
agent = create_agent(
    model=llm,
    tools=[google_search],
    system_prompt="You are a agent who can search the web using google search"
)
```

```python
question = "what is today tech news?"
```

```python
res = agent.invoke({"messages":[{"role":"user","content":question}]})
```

```python
print(res["messages"][-1].content)
```

ADDING MEMORY TO GET ANSWER FROM PREVIOUS CONVERSATIONS


```python
from langgraph.checkpoint.memory import MemorySaver
```

```python
agent1 = create_agent(
    model=llm,
    tools=[google_search],
    checkpointer=MemorySaver(),
    system_prompt="You are a agent who caan search the web using google search and also tell use about our previous conversations"
)
```

```python
question = "tell me about yesterday's "
```

```python
res = agent1.invoke(
    {"messages":[{"role":"user","content":question}]},
    {"configurable":{"thread_id":"google_search_agent_thread"}}
)
```

```python
print(res["messages"][-1].content)
```
