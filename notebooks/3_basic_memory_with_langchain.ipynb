```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(model="gpt-4o")
```

```python
prompts = [
    {"role":"user" , "content":"Hello , My name is Ayush"},
    {"role":"ai" , "content" : "Hello Ayush, How can I help you?"},
    {"role":"user" , "content":"Tell me my name?"}
]

res = llm.invoke(prompts)
```

```python
res.content
```

Manage History

```python
history = []
while True:
    query = input("User: ")

    if query.lower() in ["quit" , "exit" , "bye"]:
        print("GoodBye 👋")
        break
    history.append({"role":"user","content":query})
    print("User: ",query)

    res = llm.invoke(history)
    history.append({"role":"ai","content":res.content})
    print("AI: ", res.content)
```
